Recreate usefull table 

```python
col = ['communityName',	'state','population','racepctblack', 'racePctWhite','racePctAsian',	'racePctHisp','ViolentCrimesPerPop','nonViolPerPop','agePct12t21',
        'agePct12t29',
        'agePct16t24',
        'agePct65up',
        'PctUnemployed',
        'PctEmploy',
        'PctEmplManu',
        'PctEmplProfServ',
        'PctOccupManu',
        'PctOccupMgmtProf']

crime = original [col]



def cleaner(race_crime):

    change = {
        'communityName': 'community_name',
        'state': 'state',
        'population': 'population',
        'racepctblack': 'race_black',
        'racePctWhite': 'race_white',
        'racePctAsian': 'race_asian',
        'racePctHisp': 'race_hisp',
        'ViolentCrimesPerPop': 'violent_crimes_per_pop',
        'nonViolPerPop': 'non_viol_per_pop',
        'agePct12t21': '12_21Y',
        'agePct12t29': '12_29Y',
        'agePct16t24': '12_24Y',
        'agePct65up': '65Y',
        'PctUnemployed': 'Unemployed',
        'PctEmploy': 'Employ',
        'PctEmplManu': 'Manu',
        'PctEmplProfServ': 'prof_service',
        'PctOccupManu': 'Occup_Manu',
        'PctOccupMgmtProf': 'Occup Mgmt Prof'
    }

    race_crime = race_crime.rename(columns=change)

    race_crime.columns = race_crime.columns.str.strip().str.lower().str.replace(' ', '_')


    race_crime['community_name'] = race_crime['community_name'].str.replace('ship', '').str.replace('city', '')

    return race_crime


crime = cleaner(crime)

crime


```

Location and criminality 

```python
#Top 5 des villes aux crimes les plus violents

top_5 = crime.sort_values(by='violent_crimes_per_pop', ascending=False)

top_5 = top_5.head(10)

display(top_5[['community_name','population']])

plt.figure(figsize=(8, 6))
plt.bar(top_5['community_name'], top_5['violent_crimes_per_pop'], color='skyblue')
plt.xlabel('community')
plt.ylabel('crimes/population')
plt.title('Top 10 most violent communoties')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

#Top 5 des villes aux crimes sans violence

top_5 = crime.sort_values(by='non_viol_per_pop', ascending=False)

top_5 = top_5.head(10)

plt.figure(figsize=(8, 6))
plt.bar(top_5['community_name'], top_5['non_viol_per_pop'], color='red')
plt.xlabel('community')
plt.ylabel('crimes/population')
plt.title('Top 10 communities with non violence crimes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
state = crime.groupby('state')['violent_crimes_per_pop'].mean()
top_5 = state.sort_values(ascending=False).head(10)

plt.figure(figsize=(9, 6))
plt.bar(top_5.index, top_5.values, color='yellow')
plt.xlabel('State')
plt.ylabel('Average crimes for 100 000 pers')
plt.title('Top 10 of the most violence States')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




state = crime.groupby('state')['non_viol_per_pop'].mean()
top_5 = state.sort_values(ascending=False).head(10)

plt.figure(figsize=(9, 6))
plt.bar(top_5.index, top_5.values, color='purple')
plt.xlabel('State')
plt.ylabel('Average crimes for 100 000 pers')
plt.title('Top 10 of States with non violence crimes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



```

```python
crime['black_population'] = crime ['race_black']/100 * crime['population']
crime['white_population'] = crime ['race_white']/100 * crime['population']
crime['asian_population'] = crime ['race_asian']/100 * crime['population']
crime['hisp_population'] = crime ['race_hisp']/100 * crime['population']

# obs_state = crime.pivot_table(values='['race_black'/','race_white', 'race_asian', 'race_hisp', index='state', columns=['race_black','race_white', 'race_asian', 'race_hisp'], aggfunc='mean')

table_pivot_race = crime.pivot_table(values=['white_population', 'black_population', 'hisp_population', 'asian_population'],index='state',aggfunc=lambda x: (x.sum() / crime.loc[x.index, 'population'].sum()) * 100)

table_pivot_crime = crime.pivot_table(values=['violent_crimes_per_pop', 'non_viol_per_pop'],
                                      index='state',
                                      aggfunc='mean')

table_pivot_race = table_pivot_race.merge(table_pivot_crime, on='state')

table_pivot_race.sort_values(by='violent_crimes_per_pop', ascending=False)


table_pivot_race.head()






```

```python

import seaborn as sns



table_pivot_race_sorted = table_pivot_race.sort_values(by='violent_crimes_per_pop', ascending=False)

tableau_resultat = table_pivot_race_sorted[['violent_crimes_per_pop', 'white_population', 'black_population', 'hisp_population', 'asian_population']]
tableau_resultat = tableau_resultat.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=tableau_resultat.index, y='violent_crimes_per_pop', data=tableau_resultat, color='grey', label='Violent Crimes')
sns.barplot(x=tableau_resultat.index, y='white_population', data=tableau_resultat, color='blue', label='White Population')
sns.barplot(x=tableau_resultat.index, y='black_population', data=tableau_resultat, color='red', label='Black Population')
sns.barplot(x=tableau_resultat.index, y='hisp_population', data=tableau_resultat, color='orange', label='Hispanic Population')
sns.barplot(x=tableau_resultat.index, y='asian_population', data=tableau_resultat, color='green', label='Asian Population')

plt.xlabel('State')
plt.ylabel('Percentage / Crimes per Population')
plt.title('Most Violent States and Ethnic Distribution')
plt.legend()
plt.show()
```

```python
race = ['white_population', 'black_population', 'hisp_population', 'asian_population']
print("For violent crime:")
for col in race:
    corr =  table_pivot_race[col].corr(table_pivot_race['violent_crimes_per_pop'])
    print('For',col, "The Pearson correlation is ",corr)

print("-------------------------------------------------------------------------------------")

for col in race:
    corr = table_pivot_race[col].corr(table_pivot_race['violent_crimes_per_pop'], method='spearman')

    print('For',col, "The Spearman correlation is ",corr)




print("                                                                                                                                                       ")

print("For non violent crime:")
for col in race:
    corr = table_pivot_race[col].corr(table_pivot_race['non_viol_per_pop'])
    print('For',col, "The Pearson correlation is ",corr)

print("-------------------------------------------------------------------------------------")

for col in race:
    corr = table_pivot_race[col].corr(table_pivot_race['non_viol_per_pop'], method='spearman')

    print('For',col, "The Spearman correlation is ",corr)
```

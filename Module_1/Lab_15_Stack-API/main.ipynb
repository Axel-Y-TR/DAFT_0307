#### Question 1: Find the questions and answers of last month.

```python
# your code here
SITE = StackAPI('stackoverflow')
questions = SITE.fetch('questions', fromdate=1686487441, todate=1689079441)

print(questions)
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here

most_voted = SITE.fetch('questions', fromdate=1686487441, todate=1689079441, tagged='python',min = 5, sort='votes')

print(most_voted)
```
